```
product_name_hp['product_name'] = product_name_hp.product_name.str.lower()

product_name_hp.head()
```

```
remove_words = ['garansi', 'tahun', '-', 'resmi', 'tam',
                'hp', 'smartphone', 'smart phone', 'smart',
                'phone', 'hape', 'hand', 'phone', 'distributor', 
                'ibox', 'gold', 'gray', 'grey', 'emas',
                'biru', 'blue', 'red', 'pink', 'silver',
                'hitam', 'black', 'new', '2017', '2016', 
                '2018', '2019', 'global', 'clear', '&',
                'carbon', 'glass', 'jelly', 'for', 'ultrathin',
                '|', 'thin', 'full', 'baby', 'mirror', 
                'power', 'slim', 'lcd', 'hard', '+',
                'custom', 'room', 'free', 'ori', 'blue',
                'in', 'touchscreen', 'tpu', 'perdana', 'batre'
                'putih', '/', 'simpati']
```

```
corpus = []
for sent in product_name_hp['product_name'][:][:]:
    tmp = sent.split(' ')
    st = []
    for w in tmp: 
        if w not in remove_words: 
            st.append(w)
    corpus.append(st)
```

```
model_en.build_vocab(corpus, update=True)
```

```
model_en.train(sentences=corpus, total_examples = len(corpus), epochs=5)
```

```
model_en.most_similar('powerbank')
```

```
import numpy as np

def sent_vectorizer(sent, model):
    sent_vec =[]
    numw = 0
    for w in sent:
        try:
            if numw == 0:
                sent_vec = model[w]
            else:
                sent_vec = np.add(sent_vec, model[w])
            numw+=1
        except:
            pass
     
    return np.asarray(sent_vec) / numw
```
