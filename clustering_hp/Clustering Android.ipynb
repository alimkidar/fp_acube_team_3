```python
import pandas as pd
import numpy as np
from numpy import random
import sklearn.cluster as cluster
from pyclustering import cluster as pycluster
from pyclustering.cluster.kmeans import kmeans, kmeans_visualizer
import sklearn
from pyclustering.cluster.center_initializer import kmeans_plusplus_initializer

def kmeans_silhouette(X, range_k) :
    silhouette = []
    for k in range_k:
        clust_result = cluster.k_means(X,k)
        centers = clust_result[0]
        category = clust_result[1]
        silhouette = silhouette + [sklearn.metrics.silhouette_score(X,category)]
    return [range_k[silhouette.index(max(silhouette))], max(silhouette)]

def ward_silhouette(X, range_k) :
    silhouette = []
    for k in range_k:
        clust_result = cluster.hierarchical.AgglomerativeClustering(n_clusters=k, linkage='ward')
        clust_result.fit(X)
        category = clust_result.labels_
        silhouette = silhouette + [sklearn.metrics.silhouette_score(X,category)]
    return [range_k[silhouette.index(max(silhouette))], max(silhouette)]

def complete_silhouette(X, range_k) :
    silhouette = []
    for k in range_k:
        clust_result = cluster.hierarchical.AgglomerativeClustering(n_clusters=k,linkage='complete')
        clust_result.fit(X)
        category = clust_result.labels_
        silhouette = silhouette + [sklearn.metrics.silhouette_score(X,category)]
    return [range_k[silhouette.index(max(silhouette))], max(silhouette)]

def average_silhouette(X, range_k) :
    silhouette = []
    for k in range_k:
        clust_result = cluster.hierarchical.AgglomerativeClustering(n_clusters=k,linkage='average')
        clust_result.fit(X)
        category = clust_result.labels_
        silhouette = silhouette + [sklearn.metrics.silhouette_score(X,category)]
    return [range_k[silhouette.index(max(silhouette))], max(silhouette)]

def single_silhouette(X, range_k) :
    silhouette = []
    for k in range_k:
        clust_result = cluster.hierarchical.AgglomerativeClustering(n_clusters=k,linkage='single')
        clust_result.fit(X)
        category = clust_result.labels_
        silhouette = silhouette + [sklearn.metrics.silhouette_score(X,category)]
    return [range_k[silhouette.index(max(silhouette))], max(silhouette)]

def shift_silhouette(X) :
    est_b = cluster.estimate_bandwidth(X)
    clust_result = cluster.mean_shift(X, est_b)
    category = clust_result[1]
    silhouette = [sklearn.metrics.silhouette_score(X,category)]
    return [len(clust_result[0]), silhouette]
def get_merk(x):
    a = x.split()
    return a[0]
def jadul(x):
    if x == '-':
        a = 1
    else: a = 0
    return a
```
